```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 설정
x = np.array([30, 45, 60, 75, 90, 105, 120])  # 운동 시간(단위: 분)
y = np.array([200, 300, 400, 500, 600, 700, 800])  # 각 운동 시간에 해당하는 소모 칼로리

# 데이터 분포 시각화
plt.figure(figsize=(6, 3))
plt.scatter(x, y)
plt.xlabel('Exercise Time (minutes)')
plt.ylabel('Calories Burned (kcal)')
plt.title('Exercise Data: Time vs Calories Burned')
plt.show()

# 초기 파라미터(weight, bias) 설정
w = 0
b = 0


# 학습률 및 에폭 설정
# 기존에는 학습률이 지나치게 커서 gradient exploding 경향이 발생하였다.
# 이를 방지하기 위해서 학습률을 0.0001로 감소시켰다.
# 또한 감소한 학습률로 인해 기존의 에폭으로는 weight와 bias가 수렴하지 않아서 이를 방지하기 위해서 에폭을 100으로 증가시켰다.
lr = 0.0001 
epochs = 100

# 데이터 개수
n = len(x)

# 경사 하강법 시작
for i in range(epochs):
    y_pred = w * x + b  # 예측값 계산
    error = y - y_pred  # 오차 계산

    # 오차 함수의 w, b에 대한 편미분
    w_diff = -(2 / n) * sum(x * (error))
    b_diff = -(2 / n) * sum(error)

    # 파라미터 업데이트
    w -= lr * w_diff
    b -= lr * b_diff

    if i % (epochs / 10) == 0:  # 실행 중 Epoch 별 출력해보기
        print(f"Epoch {i}: w = {w:.2f}, b = {b:.2f}")

# 최종 모델 시각화
plt.figure(figsize=(6, 3))
plt.scatter(x, y, color='blue')
plt.plot(x, w * x + b, color='red')
plt.xlabel('Exercise Time (minutes)')
plt.ylabel('Calories Burned (kcal)')
plt.title('Final Model: Time vs Calories Burned')
plt.show()

print(f"Final model parameters: w = {w:.2f}, b = {b:.2f}")
```
